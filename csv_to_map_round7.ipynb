```python
import pandas as pd

election_result = pd.read_csv('./round07_election_result2.csv')
df = pd.DataFrame(election_result, columns=['광역시도', '시군', 'pop', 'minju', 'hanguk', 'gita'])
# df

df_special = df[df['광역시도'].isin(['광주광역시','전라남도'])]

df_special['gita'] += df_special['hanguk']
df_special['hanguk'] = 0
# df_special

df[df['광역시도'].isin(['광주광역시','전라남도'])] = df_special
df[df['광역시도'].isin(['광주광역시','전라남도'])]


# df['minju_rate'] = df['minju'] / df['pop']
# df['hanguk_rate'] = df['hanguk'] / df['pop']
# df
```

```python
import pandas as pd

df_align = df.groupby(df['광역시도']).agg({'pop':sum, 'minju':sum, 'hanguk':sum,
                           'gita':sum})
df_align['minju_rate'] = df_align['minju'] / df_align['pop']
df_align['hanguk_rate'] = df_align['hanguk'] / df_align['pop']
df_align['gita_rate'] = df_align['gita'] / df_align['pop']
df_align
```

```python
df_align.to_csv('round07_election_result_aligned.csv',
               sep=',')
```
