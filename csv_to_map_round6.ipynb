```python
import pandas as pd

election_result = pd.read_csv('./round06_election_result.csv')
df = pd.DataFrame(election_result, columns=['광역시도', '시군', 'pop', 'saenuri', 'minju', 'gita'])
# df

special_do = ['부산광역시','울산광역시']
df_special = df[df['광역시도'].isin(special_do)]

df_special['gita'] += df_special['minju']
df_special['minju'] = 0
# df_special

df[df['광역시도'].isin(special_do)] = df_special
df
```

```python
df_align = df.groupby(df['광역시도']).agg({'pop':sum, 'minju':sum, 'saenuri':sum,
                           'gita':sum})
df_align['minju_rate'] = df_align['minju'] / df_align['pop']
df_align['saenuri_rate'] = df_align['saenuri'] / df_align['pop']
df_align['gita_rate'] = df_align['gita'] / df_align['pop']
df_align
```

```python
df_align.to_csv('round06_election_result_aligned.csv',
               sep=',')
```
