```python
import pandas as pd

election_result = pd.read_csv('./round06_election_result.csv')
df = pd.DataFrame(election_result, columns=['광역시도', '시군', 'pop', 'saenuri', 'minju', 'gita'])
df
```

```python
df_special = df[df['광역시도'].isin(['광주광역시','전라남도'])]

df_special['gita'] += df_special['saenuri']
df_special['saenuri'] = 0
df_special
```

```python
df[df['광역시도'].isin(['광주광역시','전라남도'])] = df_special
```

```python
df[df['광역시도'].isin(['광주광역시','전라남도'])]
```

```python
df['saenuri_rate'] = df['saenuri'] / df['pop']
df['minju_rate'] = df['minju'] / df['pop']
df
```

```python
df['시군'].unique()
```

```python
si_name = [None] * len(df)
# print(si_name)
              
tmp_gu_dict = {'수원':['수원시장안구', '수원시권선구', '수원시팔달구', '수원시영통구'], 
               '성남':['성남시수정구', '성남시중원구', '성남시분당구'], 
               '안양':['안양시만안구', '안양시동안구'], 
               '안산':['안산시상록구', '안산시단원구'], 
               '고양':['고양시덕양구', '고양시일산동구', '고양시일산서구'], 
               '용인':['용인시처인구', '용인시기흥구', '용인시수지구'], 
               '청주':['청주시상당구', '청주시서원구', '청주시흥덕구', '청주시청원구'], 
               '천안':['천안시동남구', '천안시서북구'], 
               '전주':['전주시완산구', '전주시덕진구'], 
               '포항':['포항시남구', '포항시북구'], 
               '창원':['창원시의창구', '창원시성산구', '창원시진해구', '창원시마산합포구', '창원시마산회원구'], 
               '부천':['부천시오정구', '부천시원미구', '부천시소사구']}

for n in df.index:
    print(df['광역시도'][n][:2], " ", df['시군'][n])
    if df['광역시도'][n][-3:] not in ['광역시', '특별시', '자치시']:
        if df['시군'][n][:-1] == '고성' and df['광역시도'][n] == '강원도':
            si_name[n] = '고성(강원)'
        elif df['시군'][n][:-1] == '고성' and df['광역시도'][n] == '경상남도':
            si_name[n] = '고성(경남)'
        else:
#             if len(df['시군'][n]) > 3:
#                 print(df['광역시도'][n], " ", df['시군'][n][:2])
#                 si_name[n] = df['시군'][n][:2] + ' ' + df['시군'][n][-3:-1]
#             else:
            si_name[n] = df['시군'][n][:-1]
            
        for keys, values in tmp_gu_dict.items():
            if df['시군'][n] in values:
                if len(df['시군'][n]) == 5:
#                     print("5", df['시군'][n][:2], " ", df['시군'][n][3:])
                    si_name[n] = df['시군'][n][:2] + ' ' + df['시군'][n][3:]
                elif len(df['시군'][n]) == 7:
#                     print("7", df['시군'][n][:2], " ", df['시군'][n][3:6])
                    si_name[n] = df['시군'][n][:2] + ' ' + df['시군'][n][3:6]
                elif df['시군'][n] in ['창원시마산합포구', '창원시마산회원구']:
#                     print("마산", df['시군'][n][:2], " ", df['시군'][n][5:-1])
                    si_name[n] = df['시군'][n][:2] + ' ' + df['시군'][n][5:-1]
                else:
#                     print("else", df['시군'][n][:2], " ", df['시군'][n][3:-1])
                    si_name[n] = df['시군'][n][:2] + ' ' + df['시군'][n][3:-1]
            
    elif df['광역시도'][n] == '세종특별자치시':
        si_name[n] = '세종'
    
    else:
        if len(df['시군'][n]) == 2:
#             print(df['광역시도'][n][:2], " ", df['시군'][n])
            si_name[n] = df['광역시도'][n][:2] + ' ' + df['시군'][n]
        else:
#             print(df['광역시도'][n][:2], " ", df['시군'][n][:-1])
            si_name[n] = df['광역시도'][n][:2] + ' ' + df['시군'][n][:-1]
        
si_name
```

```python
df['ID'] = si_name
```

```python
df_folium = df.set_index('ID')
df_folium.head()
```
