```python
# 세종 x
import pandas as pd

election_result = pd.read_csv('./round05_election_result.csv')
df = pd.DataFrame(election_result, columns=['광역시도', '시군', 'pop', 'hannara', 'minju', 'gita'])
# df

special_do1 = ['제주특별자치도']
special_do2 = ['경상남도']

df_special1 = df[df['광역시도'].isin(special_do1)]
df_special1['gita'] += df_special1['hannara']
df_special1['hannara'] = 0

df_special2 = df[df['광역시도'].isin(special_do2)]
df_special2['gita'] += df_special2['minju']
df_special2['minju'] = 0
# # df_special

df[df['광역시도'].isin(special_do1)] = df_special1
df[df['광역시도'].isin(special_do2)] = df_special2
df
```

```python
df_align = df.groupby(df['광역시도']).agg({'pop':sum, 'minju':sum, 'hannara':sum,
                           'gita':sum})
df_align['minju_rate'] = df_align['minju'] / df_align['pop']
df_align['hannara_rate'] = df_align['hannara'] / df_align['pop']
df_align['gita_rate'] = df_align['gita'] / df_align['pop']
df_align
```
